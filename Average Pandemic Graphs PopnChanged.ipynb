```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import time
```

```python
#PopnChanged R = 4
z = 1.960
test = pd.read_csv (r'E:\Users\Michael\Project Notebook\Sim Data\PopnChanged\4\PopnChanged_R4_1.csv')
t = np.zeros((100,len(test)))
S = np.zeros((100,len(test)))
E = np.zeros((100,len(test)))
I = np.zeros((100,len(test)))
R = np.zeros((100,len(test)))
D = np.zeros((100,len(test)))
V = np.zeros((100,len(test)))
V_c = np.zeros((100,len(test)))
T = np.zeros((100,len(test)))

t_avg = np.zeros(len(test))
S_avg = np.zeros(len(test))
E_avg = np.zeros(len(test))
I_avg = np.zeros(len(test))
R_avg = np.zeros(len(test))
D_avg = np.zeros(len(test))
V_avg = np.zeros(len(test))
V_c_avg = np.zeros(len(test))
T_avg = np.zeros(len(test))

t_Err = np.zeros(len(test))
S_Err = np.zeros(len(test))
E_Err = np.zeros(len(test))
I_Err = np.zeros(len(test))
R_Err = np.zeros(len(test))
D_Err = np.zeros(len(test))
V_Err = np.zeros(len(test))
V_c_Err = np.zeros(len(test))
T_Err = np.zeros(len(test))
for i in range(1,101):
    df = pd.read_csv (r'E:\Users\Michael\Project Notebook\Sim Data\PopnChanged\4\PopnChanged_R4_{}.csv'.format(i))
    t[i-1,:] = df['t']
    S[i-1,:] = df['S']
    E[i-1,:] = df['E']
    I[i-1,:] = df['I']
    R[i-1,:] = df['R']
    D[i-1,:] = df['D']
    V[i-1,:] = df['Vaccinated']
    V_c[i-1,:] = df['Cumulative Vaccinated']
    T[i-1,:] = df['Total Cases']

t_avg = np.mean(t, axis = 0)
S_avg = np.mean(S, axis = 0)
E_avg = np.mean(E, axis = 0)
I_avg = np.mean(I, axis = 0)
R_avg = np.mean(R, axis = 0)
D_avg = np.mean(D, axis = 0)
V_avg = np.mean(V, axis = 0)
V_c_avg = np.mean(V_c, axis = 0)
T_avg = np.mean(T, axis = 0)

t_Err = z*np.std(t, axis = 0)/np.sqrt(100)
S_Err = z*np.std(S, axis = 0)/np.sqrt(100)
E_Err = z*np.std(E, axis = 0)/np.sqrt(100)
I_Err = z*np.std(I, axis = 0)/np.sqrt(100)
R_Err = z*np.std(R, axis = 0)/np.sqrt(100)
D_Err = z*np.std(D, axis = 0)/np.sqrt(100)
V_Err = z*np.std(V, axis = 0)/np.sqrt(100)
V_c_Err = z*np.std(V_c, axis = 0)/np.sqrt(100)
T_Err = z*np.std(T, axis = 0)/np.sqrt(100)



plt.style.use("ggplot")
sline = plt.plot(t_avg,S_avg,color="red",linewidth=2,label = 'Susceptible')
sline_err = plt.fill_between(t_avg, (S_avg-S_Err), (S_avg+S_Err), color = "r",alpha=.1)
eline = plt.plot(t_avg,E_avg,color="yellow",linewidth=2,label = 'Exposed')
eline_err = plt.fill_between(t_avg, (E_avg-E_Err), (E_avg+E_Err), color = "y",alpha=.1)
iline = plt.plot(t_avg,I_avg,color="green",linewidth=2,label = 'Infected')
iline_err = plt.fill_between(t_avg, (I_avg-I_Err), (I_avg+I_Err), color = "g",alpha=.1)
rline = plt.plot(t_avg,R_avg,color="blue",linewidth=2,label = 'Recovered')
rline_err = plt.fill_between(t_avg, (R_avg-R_Err), (R_avg+R_Err), color = "b",alpha=.1)
dline = plt.plot(t_avg,D_avg,color="purple",linewidth=2,label = 'Dead')
dline_err = plt.fill_between(t_avg, (D_avg-D_Err), (D_avg+D_Err), color = "purple",alpha=.1)
tline = plt.plot(t_avg,T_avg,color="orange",linewidth=2,label = 'Total Cases')
tline_err = plt.fill_between(t_avg, (T_avg-T_Err), (T_avg+T_Err), color = "orange",alpha=.1)
vline = plt.plot(t_avg,V_avg,color="teal",linewidth=2,label = 'Vaccinated')
vline_err = plt.fill_between(t_avg, (V_avg-V_Err), (V_avg+V_Err), color = "teal",alpha=.1)
vcline = plt.plot(t_avg,V_c_avg,color="lawngreen",linewidth=2,label = 'Cumulative Vaccinated')
vcline_err = plt.fill_between(t_avg, (V_c_avg-V_c_Err), (V_c_avg+V_c_Err), color = "lawngreen",alpha=.1)
plt.xlabel("Time (days)",fontweight="bold")
plt.ylabel("Number of people",fontweight="bold")
legend = plt.legend(title="Population",loc=5,bbox_to_anchor=(1.50,0.5))
frame = legend.get_frame()
frame.set_facecolor("white")
frame.set_linewidth(0)
```
