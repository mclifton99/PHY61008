```python
plt.style.use("ggplot")
caseline = plt.plot("Day","Total Cases","",data=Data,color="blue",linewidth=2)
caseline1 = plt.plot("Day","London Cases","",data=Data,color="red",linewidth=2)
caseline2 = plt.plot("Day","Birmingham Cases","",data=Data,color="maroon",linewidth=2)
caseline3 = plt.plot("Day","Leeds Cases","",data=Data,color="black",linewidth=2)

caseline5 = plt.plot("Day","Sheffield Cases","",data=Data,color="cyan",linewidth=2)
caseline6 = plt.plot("Day","Manchester Cases","",data=Data,color="fuchsia",linewidth=2)
caseline7 = plt.plot("Day","Bradford Cases","",data=Data,color="salmon",linewidth=2)

caseline9 = plt.plot("Day","Liverpool Cases","",data=Data,color="purple",linewidth=2)
caseline10 = plt.plot("Day","Bristol Cases","",data=Data,color="orange",linewidth=2)


plt.xlabel("Time (days)",fontweight="bold")
plt.ylabel("Number of people",fontweight="bold")
legend = plt.legend(title="Population",loc=5,bbox_to_anchor=(1.50,0.5))
frame = legend.get_frame()
frame.set_facecolor("white")
frame.set_linewidth(0)
```
